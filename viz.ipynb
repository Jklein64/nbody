```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.array([
    4000,
0.030235,
1000,
0.002376,
2000,
0.007351,
4000,
0.028701,
8000,
0.112463,
16000,
0.449947,
32000,
1.784391,
64000,
7.242116
])
```

```python
data = data.reshape((-1, 2))[1:]
```

```python
data
```

```python
import scipy as sp
plt.plot(data[:, 0], data[:, 1], label="true runtime")
(a, b, c), _ = sp.optimize.curve_fit(lambda x, a, b, c: a*x**2 + b*x + c, data[:, 0], data[:, 1])
x = np.linspace(np.min(data[:, 0]), np.max(data[:, 0]), 100)
plt.plot(x, a*x**2 + b*x + c, alpha=.5, ls="dotted", lw=2, label="quadratic curve")
plt.title("runtime vs particle count (naive)")
plt.ylabel("runtime (s)")
plt.xlabel("particle count")
plt.legend()
```
